```python
import numpy as np
import pandas as pd
#import matplotlib.pylot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
airline = pd.read_csv("/content/data set arima1.csv", index_col ='Date',parse_dates = True)
```

```python
airline.head()
  
```

```python
result = seasonal_decompose(airline['Nifty 50'],model ='multiplicative')
  
# ETS plot 
result.plot()
```
